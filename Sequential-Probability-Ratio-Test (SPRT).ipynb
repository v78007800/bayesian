```python
# SPRT 
# Binomial Endpoints
import sprt as sprt
import numpy as np

# веростность сбоя была 0.45
h0 = 0.45
# мы не хотим чтоб она выросла выше 0.5
h1 = 0.5
# Если сбои реально не врыосли, вероятность ошибочного утверждения о том, что они выросли, должна быть меньше, чем 0.05
alpha = 0.05
# Type II error rate = 0.2
beta = 0.2
# генерируем данные
values = np.random.binomial(1, 0.5, 300)
print('Среднее ',values.mean())
test = sprt.SPRTBinomial(h0 = h0, h1 = h1, alpha = alpha, beta = beta, values = values)
```

```python
#если пробиваем верхнюю границу, значит сбои перевалили кретичный порог, если нижний - то отметаем гипотезу о росте сбоев
test.plot() #test.plotBoundary() #test.plot(fill = False)
```
